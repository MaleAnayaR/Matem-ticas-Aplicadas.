#Ejercicio 1: Cilindro inscrito en un cono

$V = \pi r^2 h$

$\frac{r}{H - h} = \frac{R}{H} \Rightarrow r = \frac{R}{H}(H - h)$

$V(h) = \pi \left( \frac{R}{H}(H - h) \right)^2 h = \pi \frac{R^2}{H^2}(H - h)^2 h$

$V'(h) = \pi \frac{R^2}{H^2} \left[ -2(H - h)h + (H - h)^2 \right]$

$= \pi \frac{R^2}{H^2}(H - h)[(H - h) - 2h] = \pi \frac{R^2}{H^2}(H - h)(H - 3h)$

Puntos críticos: $h = H$ (borde), $h = H/3$

$h = \frac{H}{3}, \quad r = \frac{R}{H}(H - H/3) = \frac{2R}{3}$

El cilindro de volumen máximo tiene altura=$H/3$ y radio=$2R/3$

### 1: Solución de ecuaciones usando el método de Newton

Para resolver las siguientes ecuaciones:

1. $$x^5 - 6x^3 + 2x = 2$$  
2. $$\sin(x) = x^2$$  
3. $$\cos(t) = t^2$$  
4. $$2\sin(x) = \sqrt{x + 1}$$  

Primero, cada ecuación se reescribió en la forma \( f(x) = 0 \), y luego se aplicó el método de Newton-Raphson con estimaciones iniciales razonables.

La fórmula utilizada fue:

$$
x_{n+1} = x_n - \frac{f(x_n)}{f'(x_n)}
$$

Para cada caso:
- Se definió \( f(x) \) y su derivada \( f'(x) \).
- Se eligió un valor inicial \( x_0 \).
- Se iteró hasta que la diferencia entre iteraciones fuera menor que $$0.0001$$.

Así se obtuvieron aproximaciones de las soluciones con el error requerido.

```python
def newton_raphson(f, fprime, x0, tol, maxiter):
    x = x0
    for i in range(maxiter):
        fx = f(x)
        if abs(fx) < tol:
            return x

        fp = fprime(x)
        if fp == 0:
            print('Error la derivada es cero')
            break
        x = x - fx / fp
    return x
```

```python
f = lambda x: x**5 - 6*x**3 + 2*x - 2
fprime = lambda x: 5*x**4 - 18*x**2 + 2

r = newton_raphson(f, fprime, 0.5, 1e-4, 100)
print(f"La raíz aproximada es: {r}")
```

### 2. Queremos aproximar $$\sqrt[3]{2}$$ usando el método de Newton-Raphson.

Sabemos que $$\sqrt[3]{2}$$ es la solución de la ecuación:

$$
x = \sqrt[3]{2} \Rightarrow x^3 = 2 \Rightarrow x^3 - 2 = 0
$$

Entonces definimos la función:

$$
f(x) = x^3 - 2
\quad \text{y su derivada:} \quad
f'(x) = 3x^2
$$

La fórmula de Newton es:

$$
x_{n+1} = x_n - \frac{f(x_n)}{f'(x_n)} = x_n - \frac{x_n^3 - 2}{3x_n^2}
$$

Repetimos esto hasta que se cumpla:

$$
|x_{n+1} - x_n| < 0.00005
$$
damos los valores en el codigo:

```python
import math

# Definimos la función y su derivada
f = lambda t: t**3 - 2
fprime = lambda t: 3 * t**2

# Valor inicial
t0 = 1.0

# Usamos el método de Newton-Raphson
def newton_raphson(fun, fun_prime, x0, tol, maxiter):
    x = x0
    for _ in range(maxiter):
        fx = fun(x)
        if abs(fx) < tol:
            return x
        fpx = fun_prime(x)
        if fpx == 0:
            print("Error: derivada es cero.")
            return None
        x = x - fx / fpx
    return x

# Llamamos el método con tolerancia 0.00005
root = newton_raphson(f, fprime, t0, 5e-5, 100)
print(f"La raíz aproximada es: {root}")
```

```python
#I.
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp

# Definir la variable simbólica
x = sp.symbols('x')

# Definir la función simbólica
h_sym = 5.235 * x**3 - 1.245 * 7.126 * x - 3.141

# Derivar la función
h_prime = sp.diff(h_sym, x)
h_double_prime = sp.diff(h_prime, x)

# Convertir a funciones numéricas
h = sp.lambdify(x, h_sym, 'numpy')
h_prime_func = sp.lambdify(x, h_prime, 'numpy')
h_double_prime_func = sp.lambdify(x, h_double_prime, 'numpy')

# Crear valores de x para graficar
x_vals = np.linspace(-3, 3, 400)
y_vals = h(x_vals)
y_prime_vals = h_prime_func(x_vals)
y_double_prime_vals = h_double_prime_func(x_vals)

# Calcular puntos críticos (raíces de la primera derivada)
critical_points = sp.solve(h_prime, x)
critical_points_real = [float(pt.evalf()) for pt in critical_points if pt.is_real]
critical_vals = [h(pt) for pt in critical_points_real]

# Calcular puntos de inflexión (raíces de la segunda derivada)
inflection_points = sp.solve(h_double_prime, x)
inflection_points_real = [float(pt.evalf()) for pt in inflection_points if pt.is_real]
inflection_vals = [h(pt) for pt in inflection_points_real]

# Graficar
plt.figure(figsize=(10, 6))
plt.plot(x_vals, y_vals, label='h(x)', color='blue')
plt.plot(x_vals, y_prime_vals, '--', label="h'(x)", color='orange')
plt.plot(x_vals, y_double_prime_vals, ':', label="h''(x)", color='green')

# Marcar puntos críticos
plt.plot(critical_points_real, critical_vals, 'ro', label='Puntos críticos')
for x_c, y_c in zip(critical_points_real, critical_vals):
    plt.text(x_c, y_c, f"({x_c:.2f}, {y_c:.2f})", fontsize=9, ha='center')

# Marcar puntos de inflexión
plt.plot(inflection_points_real, inflection_vals, 'ko', label='Puntos de inflexión')
for x_i, y_i in zip(inflection_points_real, inflection_vals):
    plt.text(x_i, y_i, f"({x_i:.2f}, {y_i:.2f})", fontsize=9, ha='center')

# Estilo del gráfico
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.title("Análisis gráfico de h(x) = 5.235x³ - 1.245·7.126x - 3.141")
plt.xlabel("x")
plt.ylabel("h(x)")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
#III.
from scipy.interpolate import CubicSpline

x_points = [2, 4, 6]
y_points = [-3, -1, 1]
cs = CubicSpline(x_points, y_points, bc_type=((1, 0.0), (1, 3.0)))
x = np.linspace(1, 8, 400)
y = cs(x)
y_prime = cs.derivative(1)(x)
y_double_prime = cs.derivative(2)(x)

plt.figure(figsize=(10, 6))
plt.plot(x, y, label="f(x)", color="blue")
plt.plot(x, y_prime, '--', label="f'(x)", color="green")
plt.plot(x, y_double_prime, '--', label="f''(x)", color="red")
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.scatter([2, 6], [-3, 1], color='purple')
plt.text(2, -3.3, "(2, -3)", ha='center')
plt.text(6, 1.2, "(6, 1)", ha='center')
plt.title("Bosquejo de una función con las condiciones dadas")
plt.xlabel("x")
plt.ylabel("f(x), f'(x), f''(x)")
plt.grid(True)
plt.legend()
plt.ylim(-5, 5)

plt.show()
```
