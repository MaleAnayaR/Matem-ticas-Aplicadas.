Debido a la naturaleza de la función, podemos concluir que entre más grande sea el valor de n dentro de la función, más cercano a 0 va a ser el límite.
los valores -1/n y 1/n tienden a 0 a medida de que n aumenta, por tanto el valor del medio tambien va a tender a 0.

$$\frac{-1} {n} \leq \frac {sin(n)} {n} \leq \frac {1} {n}$$

acotacion por funciones:

es decir que la funcion se encuentra entre eso limites, si observamos los limites de las funciones que la acotan:

$$\lim_{n \to∞} \frac{-1} {n} =0$$

$$\lim\limits_{n \to∞} \frac{1} {n} =0$$

como las dos tiende al mismo valor, la funcion que esta entre ellas tambien va a tender a ese valor: 0.

$$\lim\limits_{n \to∞} \frac{sin(n)} {n} =0$$


```python
import numpy as np
import matplotlib.pyplot as plt

n = np.linspace(1, 50, 100)

y1 = -1/n
y2 = 1/n
y_middle = np.sin(n)/n

plt.figure(figsize=(8, 5))
plt.plot(n, y1, label=r'$-\frac{1}{n}$', linestyle='dashed', color='red')
plt.plot(n, y2, label=r'$\frac{1}{n}$', linestyle='dashed', color='blue')
plt.plot(n, y_middle, label=r'$\frac{\sin(n)}{n}$', color='black')

plt.axhline(0, color='gray', linewidth=1, linestyle='dotted')
plt.xlabel("n")
plt.ylabel("Valor de la función")
plt.title("Teorema del Sándwich para $\sin(n)/n$")
plt.legend()
plt.grid()
plt.show()
```

13. dos sucesiones que divergan pero su suma convergan

$a_n$= $(-1)^nn$ (oscila alternando signos y con crecimiento)

$b_n$= $(-1)^{n+1}n$

{$a_n+b_n$}= $(-1)^n$ + $(-1)^{n+1}$ $=$ $0$

porque $a_n$ toma los siguientes valores alternante:

$a_n$={$-1,2,-3,4,-5...$}

$b_n$={$1,-2,3,-4,5...$}

y como la suma es termino a temino, es decir $a_1$ + $b_1$ y asi sucesivamente, es una suma de sucesiones divergentes que su resultado, va a ser 0 en cada termino y va a converger a 0.



```python
# Parte 1.3 Taller Aplicadas I
# Funcion para calcular la suma infinita de una serie geometrica
def suma_geometrica(a, r):
    # Si el valor absoluto de r es menor que 1, la serie tiene una suma finita
    if abs(r) < 1:
        return a / (1 - r)  # Aplicamos la formula de la suma infinita
    else:
        return "No converge"  # Si r es mayor o igual a 1, la serie no tiene suma finita

# Lista de series geometricas con sus valores iniciales (a) y razones (r)
series = [
    (1/3, 1/3),   # Serie 1
    (4, 0.21),    # Serie 2
    (2/5, 1/5),   # Serie 3
    (2, 1/7),     # Serie 4
    (7/3, 2/5),   # Serie 5
    (1, 2/3),     # Serie 6
    (1/2, 1/2),   # Serie 7
    (1, 0.99999), # Serie 8 (Converge muy cerca de 100000)
    (1, 1.00001), # Serie 9 (No converge)
    (15/7, 3/7),  # Serie 10
    (91/89, 91/89), # Serie 11 (No converge)
    (1, 3.1416/4) # Serie 12 (No converge)
]

# Recorremos cada serie en la lista y mostramos su suma si es posible calcularla
for i, (a, r) in enumerate(series, start = 1):
    print(f"El resultado de la serie {i}, es {suma_geometrica(a, r)}")
```

```python
# Punto 2
import math

# Funcion que calcula la suma parcial hasta n terminos
def suma_parcial(n):
    suma = 0
    for j in range(n + 1):
        suma += 1 / math.factorial(j)  # Se suma 1/j! en cada iteracion
    return suma

# Punto (a): Mostrar que s_n < 3 para todo n
for n in range(21):
    print(f"s({n}) = {suma_parcial(n)}")  # Imprime cada suma parcial

# Punto (b): Verificar que la serie infinita es menor que 3
print("\n¿Siempre es menor que 3?", all(suma_parcial(n) < 3 for n in range(21)))

# Punto (c): Encontrar el valor al que converge la serie
valor_real = math.e  # La serie converge a e
aproximacion = 0
n = 0

while abs(aproximacion - valor_real) > 1e-6:  # Se detiene cuando la diferencia es pequeña
    aproximacion = suma_parcial(n)
    n += 1

print(f"\nLa suma se aproxima a e usando {n} terminos: {aproximacion}")

# Punto 7
def distancia_total(altura_inicial, porcentaje_rebote):
    distancia = altura_inicial  # La pelota cae inicialmente desde la altura dada
    altura_rebote = altura_inicial * porcentaje_rebote  # Calculamos el primer rebote

    while altura_rebote > 0.0001:  # Se detiene cuando la altura del rebote es muy pequena
        distancia += 2 * altura_rebote  # Se suma la subida y la bajada del rebote
        altura_rebote *= porcentaje_rebote  # Se calcula la nueva altura del rebote

    return distancia

# Caso 1: La pelota rebota al 75% de la altura anterior
distancia_75 = distancia_total(10, 0.75)

# Caso 2: La pelota rebota al 25% de la altura anterior
distancia_25 = distancia_total(10, 0.25)

# Imprimimos los resultados
print(f"Distancia total recorrida con rebote del 75%: {distancia_75} metros")
print(f"Distancia total recorrida con rebote del 25%: {distancia_25} metros")

# Punto 11
def terminos_serie_harmonica(limite):
    suma = 0
    n = 0

    while suma <= limite:  # Se detiene cuando la suma supera el limite
        n += 1
        suma += 1 / n  # Se suma el termino 1/n

    return n

# Encontramos cuántos términos se necesitan para superar 5 y 10
terminos_5 = terminos_serie_harmonica(5)
terminos_10 = terminos_serie_harmonica(10)

# Mostramos los resultados
print(f"Se necesitan {terminos_5} terminos para que la suma sea mayor a 5.")
print(f"Se necesitan {terminos_10} terminos para que la suma sea mayor a 10.")
```
